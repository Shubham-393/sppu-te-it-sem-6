### Merge  & Concat Cheat Sheet  

#### ðŸ”— `pd.merge(df1, df2, on="key", how="...")`  
- **inner** â†’ Only matching rows (like SQL INNER JOIN)  
- **outer** â†’ All rows, fill missing with NaN (like FULL OUTER JOIN)  
- **left** â†’ All from left, matching from right (like LEFT JOIN)  
- **right** â†’ All from right, matching from left (like RIGHT JOIN)  

#### ðŸ“Œ `pd.concat([df1, df2], axis=...)`  
- **axis=0** â†’ Add rows (stack like appending data)  
- **axis=1** â†’ Add columns (side-by-side like features merging)  

ðŸš€ **Merge for relations, Concat for stacking!**





```python
merge_inner = pd.merge(subset6,subset8, left_index=True, right_index=True) #default how='inner' 
merge_inner 
```

```python
# Merging the DataFrames on Common Columns

merged_data = pd.merge(subset6,subset8,how='outer',suffixes=("_A", "_B"))
merged_data
```

```python
merged_left = pd.merge(subset6,subset8,left_index=True,right_index=True ,how='right')
merged_left
```

```python
concat_col = pd.concat([subset6,subset8], axis=1)
concat_col
```

```python
# (Concatenating Rows) Merging by Stacking Rows 
concat_row = pd.concat([subset6,subset8], axis=0)
concat_row
```
